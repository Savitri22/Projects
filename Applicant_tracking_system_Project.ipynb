```python
# import the library

import docx2txt
```

```python
# Load the resume of applicant

resume = docx2txt.process("CS_3_APR.docx")
print(resume)
```

```python
job_desc = docx2txt.process("job_description.docx")
print(job_desc)
```

```python
# Creating List of text to store resume and job_desc

text = [resume,job_desc]
```

```python
# CountVertorizer helps to count occurence of each word

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
count_matrix = cv.fit_transform(text)
```

```python
# To know how similar the text in document and resume 
# use cosine_matrix

from sklearn.metrics.pairwise import cosine_similarity
print(cosine_similarity(count_matrix))
```

```python
# The above value showing 0.61032091 
##tells how much text in both the document matches
```

```python
match = cosine_similarity(count_matrix)[0][1]  # store %ge in a variable
match = match*100
match = round(match,2)  #upto 2 decimal places
print(match)
```

```python
# condition for application whether accepted or rejected

if (match > 60 ):
    print("Hurray!! Application is Accepted!!")
else:
    print("Regret!! Application is not fulfilling company's requirements")
```
